```python
import json
import os
import pandas as pd
```

```python
try:
    with open("all_questions.json", "r", encoding='utf-8') as file:  # ðŸ‘ˆ Fix here
        questions = json.load(file)
except FileNotFoundError:
    print("Error: File not found.")
except json.JSONDecodeError as e:
    print(f"Invalid JSON: {e}")
```

```python
questions
```

```python
df = pd.DataFrame(questions)
```

```python
df
```

```python
data = []
for q in questions:
    row = {
        "Question": q["question"],
        "Difficulty": q["difficulty"],
        "Correct Answer": q["answer"],
        "Option 1": q["options"].get("A", ""),  # Use .get() to avoid KeyErrors
        "Option 2": q["options"].get("B", ""),
        "Option 3": q["options"].get("C", ""),
        "Option 4": q["options"].get("D", ""),
        "Topic": q["topic"]
    }
    data.append(row)

# Create DataFrame
df = pd.DataFrame(data)

# Reorder columns (optional)
df = df[["Question", "Difficulty", "Correct Answer", "Option 1", "Option 2", "Option 3", "Option 4", "Topic"]]
```

```python
df.head()
```

```python
df.info()
```

```python
# Create a mapping dictionary for standardization
difficulty_map = {
    'Medium': 'medium',
    'Easy': 'easy',
    'Hard': 'hard',
    'advanced': 'hard',
    'middle': 'medium',
    'area': 'hard',
    ' easy': 'easy'
    # Add any other variants you encounter
}

# Convert to lowercase first to handle case variations
df['Difficulty'] = df['Difficulty'].str.lower()

# Replace values using the mapping dictionary
df['Difficulty'] = df['Difficulty'].replace(difficulty_map)

# Verify the results
print(df['Difficulty'].value_counts())
```
